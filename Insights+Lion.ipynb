# **Regresion lineal**

Para este proyecto, decidí utilizar el método de feature selection basado en modelos para reducir la dimensionalidad de mi dataset de Udemy y poder enfocarme en las variables más relevantes para mi objetivo. Posteriormente, utilicé una regresión lineal para predecir la cantidad de suscriptores de un curso en función del precio y la categoría a la que pertenece. La variable dependiente seleccionada fue 'num_subscribers' y las variables independientes fueron 'price' y 'category'.

```python
mean_price = data_udemy['price'].mean()
print(mean_price)
```

```python
data_udemy['category'] = pd.factorize(data_udemy['category'])[0]

train_data = data_udemy.sample(frac=0.8, random_state=0)
test_data = data_udemy.drop(train_data.index)

X_train = train_data[['price', 'category']]
y_train = train_data['num_subscribers']

X_test = test_data[['price', 'category']]
y_test = test_data['num_subscribers']

ct = ColumnTransformer([('one_hot_encoder', OneHotEncoder(categories='auto', sparse=False), [1])], remainder='passthrough')

X_train = ct.fit_transform(X_train)

X_test = ct.transform(X_test)

model = LinearRegression()
model.fit(X_train, y_train)

# Predecir el número de suscriptores para los cursos en función de su precio y categoría
# Selecciona las categorías Business, Development e IT & Software (las categorías 1, 2 y 12 respectivamente)
X_pred = ct.transform([[mean_price, 1], [mean_price, 2], [mean_price, 12]])
y_pred = model.predict(X_pred)
print(y_pred)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Predecir valores de y a partir de X de entrenamiento
y_pred_train = model.predict(X_train)

# Predecir valores de y a partir de X de prueba
y_pred_test = model.predict(X_test)

# Calcular el error cuadrático medio de los valores de entrenamiento
mse_train = mean_squared_error(y_train, y_pred_train)

# Calcular el error cuadrático medio de los valores de prueba
mse_test = mean_squared_error(y_test, y_pred_test)

# Calcular el coeficiente de determinación de los valores de entrenamiento
r2_train = r2_score(y_train, y_pred_train)

# Calcular el coeficiente de determinación de los valores de prueba
r2_test = r2_score(y_test, y_pred_test)

# Imprimir los resultados
print("Error cuadrático medio (entrenamiento): {:.2f}".format(mse_train))
print("Error cuadrático medio (prueba): {:.2f}".format(mse_test))
print("Coeficiente de determinación (entrenamiento): {:.2f}".format(r2_train))
print("Coeficiente de determinación (prueba): {:.2f}".format(r2_test))
```

El resultado de la predicción en esta regresión lineal indica que el número de suscriptores esperados para un curso en la categoría de  Busines es de alrededor de 2,853, para un curso en la categoría de  Development es de alrededor de 3,087 y para un curso en la categoría de IT & Software es de alrededor de 3,522. Las predicciones se basan en un precio de $81.67, que es la media de los precios de los cursos en el conjunto de datos.


Los resultados indican que el modelo tiene un bajo rendimiento tanto en el conjunto de entrenamiento como en el conjunto de prueba. El error cuadrático medio es bastante alto en ambos conjuntos, lo que puede sugerir que el modelo tiene dificultades para predecir con precisión el número de suscriptores. Además, los coeficientes de determinación son muy bajos, lo que indica que el modelo explica una cantidad muy pequeña de la variabilidad en los datos.

Es posible que el modelo esté sufriendo de underfitting. Podría intentar agregar más características o utilizar un modelo más complejo para mejorar el rendimiento.

# **API Google Trends**

Después de considerar otras APIs públicas que pudieran complementar mi análisis de datos de los cursos de Udemy, decidí utilizar la API de Google Trends. Google Trends brinda datos sobre cómo cambia el interés de búsqueda de un término específico a lo largo del tiempo. Esto puede ayudar a entender las tendencias del mercado y la demanda de diferentes tipos de cursos, especificamente, en lo que refieres a las busquedas de los cursos de Udemy.

 Los datos muestran la popularidad relativa de cada término de búsqueda a lo largo del tiempo, en una escala de 0 a 100. Donde 100 representa el pico de popularidad del término de búsqueda durante un período de tiempo.

```python
!pip install pytrends
```

```python
from pytrends.request import TrendReq

pytrends = TrendReq(hl='en-US', tz=360)

categorias = ['Development', 'Business', 'IT & Software'] #utilizo las 3 categorias mas populares en Udemy

trends_data = {}

for categoria in categorias:
    pytrends.build_payload([categoria], timeframe='today 5-y', geo='')
    interest_over_time_df = pytrends.interest_over_time()

    trends_data[categoria] = interest_over_time_df

for categoria, trends_df in trends_data.items():
    print(f"Trends data for {categoria}:")
    print(trends_df.head())
```

Estos resultados son un conjunto de datos de las tendencias de búsqueda en Google para las categorías "Development", "Business", y "IT & Software" durante los últimos 5 años hasta la fecha actual.

```python
for categoria, trends_df in trends_data.items():
    plt.plot(trends_df.index, trends_df[categoria], label=categoria)

plt.title('Tendencias de búsqueda para categorías de cursos en Udemy')
plt.xlabel('Año')
plt.ylabel('Interés de búsqueda')
plt.legend()
plt.show()
```

```python
top_categorias = data_edX['Course Subject'].value_counts().head(10)

plt.bar(top_categorias.index, top_categorias.values)
plt.xlabel('Categoría')
plt.ylabel('Cantidad de cursos')
plt.title('Categorías de cursos más populares en edX')
plt.xticks(rotation=90)
plt.show()
```

En el dataset de Udemy, las categorías con mayor número de estudiantes inscritos son 'Development', 'Business' y 'IT & Software'. Esto indica que hay una gran demanda y popularidad para cursos relacionados con el desarrollo de habilidades técnicas, negocios y tecnología.

Estas categorías reflejan un interés en áreas de estudio más amplias, que abarcan humanidades, ciencias sociales, ciencias naturales y ciencias aplicadas.

El análisis de ambos datasets nos muestra diferencias en las preferencias y enfoques de los estudiantes en cuanto a las áreas de estudio.

Mientras que en Udemy las categorías se centran más en habilidades técnicas y de negocios, en edX se observa un interés más diversificado en diferentes campos académicos y disciplinas.

Estos resultados resaltan la importancia de comprender las preferencias y necesidades de los estudiantes al diseñar estrategias educativas y programas de formación en línea.

```python
cursos_auditados = data_edX['Audited (> 50% Course Content Accessed)']
cursos_certificados = data_edX['% Certified of > 50% Course Content Accessed']
participantes = data_edX['Participants (Course Content Accessed)']

plt.scatter(cursos_auditados, cursos_certificados, s=participantes/1000, alpha=0.5)

plt.xlabel('Cantidad de Cursos Auditados')
plt.ylabel('Porcentaje de Cursos Certificados')
plt.title('Relación entre Cursos Auditados, Certificados y Número de Estudiantes')

plt.legend(handles=[plt.scatter([], [], s=10, label='Número de Estudiantes (en miles)')])
plt.gca().legend(loc='upper right')

plt.show()
```

El gráfico de dispersión muestra la relación entre la cantidad de cursos auditados, el porcentaje de cursos certificados y el número de estudiantes en el dataset de Harvard y MIT en la plataforma edX. Por otro lado, en el dataset de data_udemy se observa una relación entre el número de inscritos y el precio de los cursos.

A medida que aumenta la cantidad de cursos auditados, no se muestra una clara tendencia en cuanto al porcentaje de cursos certificados. Sin embargo, se puede apreciar que hay una concentración de cursos con una cantidad relativamente alta de estudiantes.

En el dataset de data_udemy, se muestra una tendencia diferente en relación al número de inscritos y el precio de los cursos. Se observa que el mayor número de inscritos se encuentra en el rango de precios de hasta 200 dólares.

En el caso de edX, no se evidencia una relación clara entre la cantidad de cursos auditados y el porcentaje de cursos certificados, lo que indica que hay otros factores que influyen en la certificación de los estudiantes. En el caso de data_udemy, se muestra que los cursos con precios más accesibles tienen mayor demanda y a medida que el precio aumenta, la cantidad de inscritos disminuye considerablemente.

En el dataset de Coursera, se observa una categoría de precios denominada "free" que representa la mayor cantidad de cursos. Esto indica que Coursera ofrece una amplia variedad de cursos gratuitos en su plataforma.

Por otro lado, en el dataset de Udemy, no se encuentra una categoría de precios específica denominada "free", lo que sugiere que Udemy tiene una estructura de precios diferente y puede tener menos cursos gratuitos disponibles en comparación con Coursera.

En resumen, al comparar los datos de Coursera y Udemy, se observa que Coursera se destaca por ofrecer una mayor cantidad de cursos gratuitos, mientras que Udemy ofrece una variedad de opciones de precios, con un incremento de inscritos en cursos más asequibles. Estas diferencias en la distribución de precios pueden reflejar las estrategias y enfoques comerciales de ambas plataformas, así como las preferencias y comportamientos de los estudiantes.

```python
top_categorias = data_coursera['category-subject-area'].value_counts().head(5).index

data_coursera_top_categorias = data_coursera[data_coursera['category-subject-area'].isin(top_categorias)]

idioma_categoria_top = data_coursera_top_categorias.groupby(['language', 'category-subject-area']).size().unstack()

idioma_categoria_top.plot(kind='bar', stacked=True)
plt.xlabel('Idioma')
plt.ylabel('Cantidad de cursos')
plt.title('Distribución de idiomas y categorías en Coursera (Top 5 categorías)')
plt.legend(title='Categoría')
plt.xticks(rotation=90)
plt.show()
```

Al comparar los conjuntos de datos de Udemy (data_udemy) y Coursera (data_coursera) en términos de las categorías de cursos más populares y los idiomas utilizados, se pueden observar algunas diferencias interesantes.

En el caso de Udemy, se encuentra que la categoría más popular, "Development", tiene una predominancia del idioma inglés, seguido de cerca por el portugués y el español.

Por otro lado, en el caso de Coursera, se observa que la mayoría de los cursos están en inglés, con casi 100 cursos en este idioma. Las categorías más populares en Coursera son "Cloud Computing", "Data Analysis", "Design and Product", "Physics and Astronomy" y "Support and Operations".

También se identifica una presencia de cursos en indonesio, particularmente en las categorías de "Support and Operations" y "Cloud Computing". Además, se encuentran algunos cursos en portugués, destacando nuevamente en la categoría de "Cloud Computing".

Estos hallazgos indican que hay similitudes en las categorías populares entre Udemy y Coursera.
